```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv ('DataSet.csv')
```

```python
df.columns
```

```python
df.describe()
```

```python
df['total_items'].value_counts()
```

```python
df['payment_method'].value_counts()
```

```python
dfCredCard = df[df['payment_method'] == 'credit_card']
dfCredCard.describe()
```

```python
dfDebit = df[df['payment_method'] == 'debit']
dfDebit.describe()
```

```python
dfCash = df[df['payment_method'] == 'cash']
dfCash.describe()
```

```python
plt.figure()
plt.hist (df['order_amount'], bins=10)
plt.title ('Sales Distribution')
plt.show()
```

```python
print ('Order Amount Median...')
print ('Credit Card:', dfCredCard['order_amount'].median())
print ('Debit:      ', dfDebit['order_amount'].median())
print ('Cash:       ', dfCash['order_amount'].median())
print ('Overall     ', df['order_amount'].median())
```

```python
print ('Total... Mode: ')
print (dfCredCard['total_items'].mode()[0])
print (dfDebit['total_items'].mode()[0])
print (dfCash['total_items'].mode()[0])
```
